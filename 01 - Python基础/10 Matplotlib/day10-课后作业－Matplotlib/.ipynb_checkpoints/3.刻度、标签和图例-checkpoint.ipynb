# 刻度、标签和图例

## pyplot接口

过程型（类似MATLAB）,交互式使用，含有诸如xlim、xticks和xticklabels之类的方法，分别控制图表的范围、刻度位置、刻度标签。

- 调用时不带参数，则返回当前的参数值

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import randn

```

- 调用时带参数，则设置参数值

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import randn

```

Tips:上述方法仅对Axesubplot起作用。
