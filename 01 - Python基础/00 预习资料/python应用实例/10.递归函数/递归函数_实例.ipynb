# 绘制n阶H-树

简单的递归绘图方案可以产生非常复杂的图像。递归绘图不仅与许多应用有关，而且还可以提供一个平台，帮助我们更好地理解递归程序的思想和特性，因为我们可以生动形象地观察递归图形的生成过程。

## n阶H-树介绍及递归思路：

观察下图1阶、2阶、3阶的H-树，理解递归树的生成规律：

2阶H-树绘制思路：
1. 绘制1阶H-树
2. 在1阶H-树的基础上绘制4个2阶H树，分别连接到H的四个顶点。附加条件是：2阶H树的大小减半，且居中绘制在H的四个顶点上

n阶H-树绘制思路：
1. 绘制n-1阶H-树
2. 在n-1阶H-树的基础上的每个H-树上绘制4个n阶H-树，分别连接到H的四个顶点。
3. 将步骤1进一步分解为步骤1和步骤2，直至n = 0

为了更加直观的理解上述绘制思路，我们先看下面的程序：

## 程序

```python
# 导入绘图模块
%matplotlib inline
import matplotlib.pyplot as plt

# 定义递归绘图函数
def draw(n,size,x,y):
    
    # 递归出口
    if n == 0:
        return
    
    
    # 递归模式（大问题分解为 n 层与原问题类似的小问题）
    
    ## 画一个大H
    x0 = x - size/2.0 ## 定位H的四个顶点
    x1 = x + size/2.0
    y0 = y - size/2.0
    y1 = y + size/2.0
    plt.plot([x0,x1],[y,y],'k') ##画出组成H的三条线段，这里使用不同的颜色绘制以便更直观
    plt.plot([x0,x0],[y0,y1],'g')
    plt.plot([x1,x1],[y0,y1],'r')
    
    ## 调用自身，以更新后的（x0,y0)为原点画1个小H
    draw(n-1,size/2.0,x0,y0) 

# 调用递归绘图函数（可以修改n的值来画不同阶的H-树）    
draw(3,0.5,0.5,0.5)
plt.show()
```

那么，如果我们想要在H-树的四个顶点上都按次规律扩展下一层H，只需要在draw函数中分别以四个顶点为原点调用自身四次，如下：

```python
def draw(n,size,x,y):
    
    # 递归出口
    if n == 0:
        return 
    
    
    # 递归模式（大问题分解为 n 层相似的小问题）
    
    ## 画一个大H
    x0 = x - size/2.0 ## 定位H的四个顶点
    x1 = x + size/2.0
    y0 = y - size/2.0
    y1 = y + size/2.0
    plt.plot([x0,x1],[y,y],'k') ##画出组成H的三条线段，这里使用不同的颜色绘制以便更直观
    plt.plot([x0,x0],[y0,y1],'g')
    plt.plot([x1,x1],[y0,y1],'r')
    
    ## 调用自身4次，分别在四个顶点画4个小H
    draw(n-1,size/2.0,x0,y0) 
    draw(n-1,size/2.0,x0,y1)
    draw(n-1,size/2.0,x1,y0)
    draw(n-1,size/2.0,x1,y1)
    
# 调用递归绘图函数（可以修改n的值来画不同阶的H-树）    
draw(3,0.5,0.5,0.5)
plt.show()
```
