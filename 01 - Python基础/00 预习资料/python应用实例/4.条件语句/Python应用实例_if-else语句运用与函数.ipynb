## Python实现赌场热门游戏（双骰子）:
    
    游戏要求：
    
    游戏开始时，赌家下好赌注，登记总资产，骰子投到7或11点时，赌家赢；
    
    骰子投到2、3、12点时，赌家输，其他情况不输不赢，不计入有效局数；
    
    赌注可以重新押，但是会对赌家处罚之前赌注的一半；
    
    可以选择把当前全部资产押注；
    
    还可以自动玩游戏，直到所获资产大于你的预期或者资产不足而淘汰。
    

```python
#导入random模块
import random
 
point_set = False
bet = 10
bankroll = 1000
sim_win = 0
sim_lose = 0

print( """
     欢迎来到游戏中心!
     你现在在玩掷骰子游戏,
     你的起始资金是 '$1000',
     你的开始赌注是 '$10',
     操作指令: 
             play():           开始掷骰子
             check_bankroll(): 检查你的当前资产
             all_in():         把全部资产押注
             set_bet():        押新注
             game():           查看游戏胜率
             auto():           能够自动玩游戏，直到超过你想要达到的资产！
""")
 #掷骰子操作函数
def roll():
  d1 = random.randrange(1,7)
  d2 = random.randrange(1,7)
  print ("你掷骰子的结果是：", d1, "+", d2, "=", d1+d2)
  return d1 + d2                   #返回掷两枚骰子的结果之和
   
def play():
    
  #global将变量定义为全局变量，可以通过定义为全局变量，实现在函数内部改变变量值。 
  global point_set, bankroll, point
  global sim_win, sim_lose
  #当客户的资产小于赌注时，提醒！！ 
  if bankroll < bet:
    print ("抱歉，你已经没钱下赌注了！")
    print ("""你想要获取更多资金吗?
        1: Yes
        2: No
       """)
    choice = input(">>")
    if choice == str(1):
      money = input("你想具体获取多少资金?")
      bankroll += int(money)
      print ("你现在的资产是 ", bankroll)
    if choice == str(2):
      print ("谢谢赏玩！ 下次见!")
  else:
    if not point_set:
      print
      print( "新游戏. 你的赌注是: ", bet)
     
    # 第一次投掷操作
    r = roll()
    if not point_set:
      if r in (7, 11):
        bankroll += bet
        sim_win += 1
        print ("祝贺您! 你赢了！你现在的资产是: ", bankroll)
      elif r in (2, 3, 12):
        bankroll -= bet
        sim_lose += 1
        print ("不好! 你输了! 你现在的资产: ", bankroll)
      else:
        point = r
        point_set = True
        print ("你的点数是", "[", point, "]")
    # 后续的投掷操作
    elif r == 7:
      bankroll -= bet
      sim_lose += 1
      point_set = False
      print ("you are crapped out! 你的资产是: ", bankroll)
    elif r == point:
      bankroll += bet
      sim_win += 1
      point_set = False
      print ("You made your point! 你的资产是: ", bankroll)
      
#定义押新注的函数                  
def set_bet(inp):
  global bet, bankroll, point_set
  print
  if point_set:
    print ("请注意!")
    print ("游戏已开始，如果你要重置赌注，你将失去一半的赌注！")
    prompt = input("""
      1: Yes, I am wanna reset my bet!
      2: No, I don't wanna reset my bet!
              """)
    if prompt == "1":
      point_set = False
      bankroll -= bet/2
      print ("罚金已没收.你现在的资产是: ", bankroll)
    else:
      pass
  bet = int(inp)
  print ("新赌注是：", bet)
  
#查把全部资产押注
def all_in():
    set_bet(bankroll)
    
#检查当前余额函数    
def check_bankroll():
  global bet
  print ("Your current balance is: ", bankroll)

#查看游戏等级 
def game():
  total = sim_win + sim_lose
  percent = float(sim_win)/total * 100
  print ("到现在为止，游戏你已经玩的有效局数为:  ", total) 
  print ("赢的次数 ", sim_win)
  print ("输的次数", sim_lose)
  print ("因此， 你的胜率是 %d%% !" %percent)

#自动玩游戏函数，直到达到你想要的资产
def auto():
  game_status = True
  purpose = input("你想要达到多少资产? ")
  while game_status:
    play()
    if bankroll > int(purpose) or bankroll < 0:
      game_status = False                     #当资产达到预期或者资产为o则停止游戏
  game()
  
```
