```python
names = ['A', 'B', 'C']
value = [1, 10, 100]
```

```python
plt.bar(names, value)
```

```python
plt.bar(names, value)
plt.xticks(range(len(names)), names)## Set locations and labels
plt.title('')
```

### 

```python
plt.subplot(131)#
plt.bar(names, value, color='r')
plt.xticks(range(len(names)), names)## Set locations and labels
plt.title('')
plt.subplot(235)
plt.scatter(names, value, color='y')
plt.subplot(333)
plt.plot(names, value, color='g')
```
