## 柱状图

```python
names = ['A', 'B', 'C']
value = [1, 10, 100]
```

```python
plt.bar(names, value)
```

```python
plt.bar(names, value)
plt.xticks(range(len(names)), names)## Set locations and labels
plt.title('离散数据散点图')
```

### 子图

```python
plt.subplot(131)#一行三列第一个位置
plt.bar(names, value, color='r')
plt.xticks(range(len(names)), names)## Set locations and labels
plt.title('离散数据散点图')
plt.subplot(235)
plt.scatter(names, value, color='y')
plt.subplot(333)
plt.plot(names, value, color='g')
```

```python
plt.subplot(131)
plt.bar(names, value, color='r')
plt.xticks(range(len(names)), names)## Set locations and labels
plt.title('离散数据散点图')
plt.subplot(236)
plt.scatter(names, value, color='y')
plt.subplot(333)
plt.plot(names, value, color='g')
plt.title('散点图，柱状图，折线图')
```

```python
tips = sns.load_dataset('tips')
```

# 1. 带状图-离散数据和连续数据之间的关系

```python
sns.stripplot(data=tips, x='day',y='total_bill', jitter=True)
```

```python
sns.stripplot(data=tips, x='day',y='total_bill', jitter=False)
```

# 2. 蜂群图 -

分析每天中， 午餐和晚餐的账单分布

# 3. Hue- 分组参数 

```python
# 付账时间
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
# 每天付账中性别比例
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# 每天付账中性别比例
sns.swarmplot(x='day',y='total_bill',data=tips,hue='size')
```

```python
# give size sort by total_bill
tips.groupby('size').count().sort_values('total_bill').index
```

```python
sns.swarmplot(x='size',y='total_bill',data=tips)
```

上图揭示了，pizza的不同size的基础价格

```python
tips['size'].corr(tips['total_bill'])
```

# 4. 箱线图

```python
sns.boxplot('day','total_bill',data=tips)
```

```python
sns.swarmplot('day','total_bill',data=tips)
```

```python
sns.boxplot('day','total_bill',data=tips, hue='time')
```

```python
# 多图合成
sns.violinplot('day','total_bill',data=tips)
sns.swarmplot('day','total_bill',data=tips, color='w')
```

# 单一变量估计

```python
## 离散型数据的核密度估计
sns.countplot(tips['day'])
```

```python
## 连续型数据的核密度估计

sns.distplot(tips['total_bill'])
```

# 抵消偏度

```python
sns.distplot(np.log(tips['total_bill']))
```

```python
sns.distplot(np.sqrt(tips['total_bill']))
```

#### 抽取前99.5 的数据(去除离散值的方法)
