## 柱状图绘制

```python
names=['A类型','B类型','C类型']
value=[1,10,100]
```

```python
plt.bar(range(len(names)),value)
plt.xticks(range(len(names)),names)
```

```python
plt.scatter(names,value)
```

```python
plt.scatter(range(len(names)),value)
plt.xticks(range(len(names)),names)
```

```python
plt.scatter(range(len(names)),value)
plt.xticks(range(len(names)),names)
plt.title('离散数据散点图')
```

### 子图-SubPlot

1.将一个画布进行切分(Figure)

2.将切分后的图分配到固定的位置

3.将图可以设置成固定的大小

```python
plt.figure(1)
plt.subplot(131)
plt.bar(names,value,color='r')
plt.subplot(236)
plt.scatter(names,value,color='y')
plt.subplot(333)
plt.plot(names,value,color='g')
plt.title('离散数据的柱状图、散点图、折线图')
```

## Seaborn的绘图练习

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

## 1.带状图-离散数据和连续数据之间的关系

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=True)
```

## 2.蜂群图-离散数据和连续数据之间的关系-密度排列

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
tips.head()
```

分析每天中，午餐和晚餐的账单分布

## 3.Hue-分组参数

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
#%matplotlib inline
```

```python
#	24.59	3.61	Female	No	Fri		4
```

在每天的付账人群中的性别分布

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='size')
```

```python
sns.swarmplot(x='size',y='total_bill',data=tips)
```

上图揭示了，pizza的不同size的基础价格

```python
tips['size'].corr(tips['total_bill'])
```

## 4.箱线图

```python
sns.boxplot("day","total_bill",data=tips)
```

```python
sns.swarmplot("day","total_bill",data=tips)
```

```python
sns.boxplot("day","total_bill",data=tips,hue='time')
```

如何来表示total_bill的概率分布
