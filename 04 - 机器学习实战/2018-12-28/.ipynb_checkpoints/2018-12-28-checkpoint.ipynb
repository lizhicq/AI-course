# 如何处理缺失值

1. 哪些列有缺失?
2. 缺失了多少？
3. 缺失率是多少？
4. 可否按照缺失率由大到小排列

```python
nullRate = train.isnull().sum()
```

```python
nullRate = [float(i)/len(nullRate) for i in nullRate]
```

# 一个算法的Baseline

数据采集->数据清洗->数据仓库->数据挖掘->数据标注->数据集市(训练集/评测集)->训练->评测->模型工程(Int8定点化)->

算法的Baseline：
训练集的制作->训练->评测
在形成一个Baseline之后，模型迭代的过程进步的标志是评测指标的提升。

评测指标：
Regression

MSE(Mean Squared Error)-L2 / MAE(Mean Absolute Error) -L1

Classification

Cross-Entropy(交叉熵)

信息量 - $log_2p$ - > 信息熵 -> $\Sigma (p *log_2p)$ -> 交叉熵 -> $\Sigma p*log_2 q$

Centor-Loss

### 缺失值的类型

Numerical | Categorial

#### Numerical

1. 均值填充
2. 中位数填充
3. 高频填充--高频会“稍微”少一些--连续型数据很难
4. 分布拟合填充--回归一个分布， 或者是($\mu\pm\sigma$ 范围随机)
5. 内在关系填充(根据列本身存在的意义和数据集内其他列的相关性进行构建数学映射模型进行)

#### Categorial

1. 高频填充
“男”|“女” --> 0|1
"小孩"|"d
